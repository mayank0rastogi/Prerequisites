```
print("## 1. Introduction ##")

from csv import reader
opened_file = open('potus_visitors_2015.csv')
read_file = reader(opened_file)

potus = list(read_file)
#print(potus)
potus = potus[2:]
potus

print("## 2. The Datetime Module ##")

import datetime as dt

print("## 3. The Datetime Class ##")

import datetime as dt

im_founded = dt.datetime(1911, 6, 16)
print("founding the particular element ")
print(im_founded)
man_on_The_moon = dt.datetime(1969, 7, 20, 20, 17)
print(man_on_The_moon)

print("## 5. Using Strptime to Parse Strings as Dates ##")
# The `potus` list of lists is available from
# the earlier screen where we created it
date_format = "%m/%d/%y %H:%M"
for row in potus:
    start_date = row[2]
    start_date = dt.datetime.strptime(start_date, date_format)
    row[2] = start_date

print(row[2])

print("## 6. Using Strftime to format dates ##")
#creating the dictionary

visitors_per_month = {}

for row in potus:
    month_dt = row[2]
    month_str = month_dt.strftime("%B, %Y")
    if month_str not in visitors_per_month:
        visitors_per_month[month_str] = 1
    else:
        visitors_per_month[month_str] += 1
print("visitors per month")
print(visitors_per_month)

print("## 7. The Time Class ##")

appt_times = []

for row in potus:
    appt_dt = row[2]
    appt_t = appt_dt.time()
    appt_times.append(appt_t)
print(appt_times)

print("## 8. Comparing time objects ##")

min_time = min(appt_times)
print(min_time)
max_time = max(appt_times)
print(max_time)

print("## 9. Calculations with Dates and Times ##")

dt_1 = dt.datetime(1981, 1, 31)
dt_2 = dt.datetime(1984, 6, 28)
dt_3 = dt.datetime(2016, 5, 24)
dt_4 = dt.datetime(2001, 1, 1, 8, 24, 13)
answer_1 = dt_2 - dt_1
print(answer_1)
answer_2 = dt_3 + dt.timedelta(days=56)
print(answer_2)
answer_3 = dt_4 - dt.timedelta(seconds=3600)
print(answer_3)

print("## 10. Summarizing Appointment Lengths ##")

for row in potus:
    end_date = row[3]
    end_date = dt.datetime.strptime(end_date, "%m/%d/%y %H:%M")
    row[3] = end_date
appt_lengths = {}

for row in potus:
    start_date = row[2]
    end_date = row[3]
    length = end_date - start_date
    if length not in appt_lengths:
        appt_lengths[length] = 1
    else:
        appt_lengths[length] += 1

min_length = min(appt_lengths)
print(min_length)
max_length = max(appt_lengths)
print(max_length)
```
